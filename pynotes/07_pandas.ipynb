```python
import pandas as pd
import numpy as np
```

```python
babynames = pd.read_csv('babynames.csv')
babynames.head()
```

Filter rows and sort values.

```python
babynames[babynames["Year"] == 2018].sort_values(by = "Count", ascending = False).head(5)
```

Find the value with specific pattern.

```python
starts_with_j = babynames["Name"].str.startswith('J') 
pd.DataFrame({'name' : babynames[starts_with_j]["Name"].unique()}).sample(10)
```

Use `map` to do calculate for each value. Use `sort_values by` to sort the table by a column.

```python
babyname_lengths = babynames["Name"].map(len)
babynames["name_length"] = babyname_lengths
babynames_by_length = babynames.sort_values(by = "name_length")
babynames_by_length = babynames_by_length.drop("name_length", 1)
babynames_by_length.head(5)
```

Calculate the length of string value and sort the table by index of values.

```python
babynames.loc[babynames['Name'].str.len().sort_values().index].head(5)
```

SQL query to find values

```python
babynames.query("Name == 'Jennifer'")["Count"]
```

Apply a function by group using `groupby` and `agg`

```python
def domain_size(series):
    return max(series) - min(series)
display(babynames.head())
name_domain_sizes = babynames.groupby("Name").agg(domain_size)
name_domain_sizes.sort_values("Count", ascending=False)
```

Select a subset of table:

* Label-based access with loc
    * Slicing for loc is inclusive

* Positional access with iloc
    * Based on numerical position instead of label
    * Slicing is exclusive


```python
elections.head()
```

```python
elections.loc[[0, 1, 2, 3, 4], ['Candidate','Party', 'Year']]
```

```python
elections.loc[0:4, 'Candidate':'Year']
```

```python
elections_year_index.loc[1980:1984, 'Candidate':'Party']
```

```python
display(elections.loc[0:4, 'Candidate'])
display(elections.loc[0:4, ['Candidate']])
```

```python
display(elections.loc[0, 'Candidate':'Year'])
display(elections.loc[[0], 'Candidate':'Year'])
```

```python
display(elections.iloc[:3, 2:])
display(elections.iloc[[0, 3, 5], [0, 3]])
```

`groupby` and `agg`

```python
babynames.groupby("Name").agg(max).head()
# babynames.groupby("Name").agg(max).plot()
```

```python
for i, (k, v) in enumerate(babynames.groupby("Year")):
    if i >= 3: break
    print(f"Year: {k}")
    display(v.head())
```

Find the location of the maximum value via the `idxmax`. Find the best per party.

```python
elections.groupby('Party')['%'].idxmax()
```

```python
best_per_party = elections.loc[elections.groupby('Party')['%'].idxmax()]
best_per_party.set_index('Party').sort_index().head() 
```

Use `groupby` and `drop_duplicates` to find the best per party.

```python
best_per_party2 = elections.sort_values('%').drop_duplicates(['Party'], keep='last')
best_per_party2.set_index('Party').sort_index().head()
```

Filter gives a copy of the original data frame where row r is included if the condition is satisfied. Note that filtering is done per group, not per row.

```python
elections.groupby("Year").filter(lambda sf: sf["%"].max() < 45)
```

```python
for i, (n, g) in enumerate(elections.groupby("Party")):
    print(n)
    display(g.head(2))
    if i>3: break
```

Two ways to `groupby` and operation ( `sum`, `max`, `mean`, `median`)

```python
display(elections.groupby("Year").agg(sum).head())
elections.groupby("Year").sum().head()
```

Group a data frame by multiple features. 

```python
babynames.groupby(["Year", "Sex"]).agg(sum).head(6)
```

```python
display(babynames.head(5))
babynames_pivot = babynames.pivot_table(
    index='Year', # the rows (turned into index)
    columns='Sex', # the column values
    values=['Count', 'Name'], # the field(s) to processed in each group
    aggfunc=np.max, # group operation
)
babynames_pivot.head(6)
```
